```python
import pandas as pd
```

```python
df = pd.read_html('http://games.espn.com/ffl/tools/projections?leagueId=1371999', header = 1)[1]
```

```python
df
```

```python
df['PLAYER'] = df['PLAYER, TEAM POS'].str.split(",", expand=True)[0]
```

```python
df['TEAM POS'] = df['PLAYER, TEAM POS'].str.split(", ").str.get(1)
```

```python
df['TEAM'] = df['TEAM POS'].str.split().str.get(0).str.upper()
```

```python
df['POSITION'] = df['TEAM POS'].str.split().str.get(1)
```

```python
index = 0
page_points = 100
df = pd.DataFrame()
while page_points >= 25:
    print(page_points)
    page_df = pd.read_html('http://games.espn.com/ffl/tools/projections?leagueId=1371999&startIndex=0' + str(index), header = 1)[1]
    page_points = page_df['PTS'].sum()
    
    df = df.append(page_df, ignore_index=True)
    index += 40
```
