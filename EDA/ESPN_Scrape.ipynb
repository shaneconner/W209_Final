```python
import pandas as pd
```

```python
# Read ESPN's HTML page
df = pd.read_html('http://games.espn.com/ffl/tools/projections?leagueId=1371999', header = 1)[1]
```

```python
# Example output
df
```

```python
# Retrieve league ID from user, another option is to ask user to paste their team URL and parse league ID from that
league_id = input("What is your league ID (found in hyperlink of your ESPN team page)? ")
```

```python
# Iterate through each page until the players combined score isn't meeting a threshold of points
index = 0
page_points = 100
df = pd.DataFrame()
while page_points >= 25:
    
    # Example of combined points
    print(page_points)
    
    # Hyperlink to iterate through until points are no longer meeting the criteria
    # Goes in jumps of (40) players per page
    page_df = pd.read_html('http://games.espn.com/ffl/tools/projections?leagueId=' \
                           + str(league_id) \
                           + '1371999&startIndex=0' \
                           + str(index), header = 1)[1]
    
    # Sum of player points
    page_points = page_df['PTS'].sum()
    
    # Append pages dataframe to main dataframe
    df = df.append(page_df, ignore_index=True)
    index += 40
```

```python
# Output
df
```

```python
# Player name column
df['PLAYER'] = df['PLAYER, TEAM POS'].str.split(",", expand=True)[0]
```

```python
df['TEAM POS'] = df['PLAYER, TEAM POS'].str.split(", ").str.get(1)
```

```python
# Player team column
df['TEAM'] = df['TEAM POS'].str.split().str.get(0).str.upper()
```

```python
# Player position column
df['POS'] = df['TEAM POS'].str.split().str.get(1)
```

```python
# Rename "type" column to match "owner" for better user comprehension
df = df.rename(index=str, columns={"TYPE": "OWNER"})
```

```python
# Switch points to ESPN points for later aggregating
df = df.rename(index=str, columns={"PTS": "ESPN PTS"})
```

```python
# Subset to only relevant columns
df = df[['PLAYER', 'POS', 'TEAM', 'OWNER', 'ESPN PTS']]
```

```python
# Output of cleaned data
df
```

```python
# Example of free agents (available players)
free_agents = df[df['OWNER'] == 'FA']
```

```python
free_agents
```

```python
# Example of unavailable players
unavailable = df[df['OWNER'] != 'FA']
```

```python
unavailable
```

```python
# Aggregation to determine value relative to other players
unavailable[['ESPN PTS']].groupby(unavailable['POS']).agg(['mean', 'count'])
```
